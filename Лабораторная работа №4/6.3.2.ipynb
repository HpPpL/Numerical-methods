```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 1.9, 2.1])
y = np.array([5.53, 2.7967, 2.25, 2.0157, 1.8856, 1.8027, 1.7454, 1.7033, 1.6712, 1.6458, 1.6252])

# Замена
t = 1 / x
s = y

# Функция для вычисления коэффициентов линейной регрессии методом наименьших квадратов
# Метод вручную уже был реализован в задаче 6.1.5, на этот раз воспользуемся готовым
# np.linalg.lstsq - решает задачу наименьших квадратов для системы линейных уравнений Ac=s
def linear_regression(t, s):
    A = np.vstack([t, np.ones(len(t))]).T
    coeffs = np.linalg.lstsq(A, s, rcond=None)[0]
    return coeffs

# Нахождение коэффициентов a и b
coeffs = linear_regression(t, s)
a = coeffs[1]
b = coeffs[0]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Исходные данные')

# Вычисление значений функции с найденными коэффициентами
x_vals = np.linspace(min(x), max(x), 400)
y_vals = a + b / x_vals

plt.plot(x_vals, y_vals, label=f'Приближение: y = {a:.4f} + {b:.4f}/x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Линейная регрессия: y = a + b/x')
plt.legend()
plt.show()

print(f'Найденные коэффициенты: a = {a}, b = {b}')
```
